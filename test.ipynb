```python
data4['local_start_time'] = pd.to_datetime(data4['local_start_time'], format='ISO8601')
data4['local_end_time'] = pd.to_datetime(data4['local_end_time'], format='ISO8601')
df4 = data4.sort_values(['consolidated_user_id', 'local_start_time'])
df4.head(10)
```

```python
df4['last_visit_end'] = df4.groupby('consolidated_user_id')['local_end_time'].shift(1)
df4['gap'] = (df4['local_start_time'] - df4['last_visit_end']).dt.total_seconds()
df4.gap
```

```python
#Bool pour savoir si on dmrre ou pas
df4['is_new_session'] = (df4['gap'] > 1800) | (df4['gap'].isna())
df4['session_number'] = df4.groupby('consolidated_user_id')['is_new_session'].cumsum()
df4['session_id'] = df4['consolidated_user_id'].astype(str) + "_" + df4['session_number'].astype(str)
```
